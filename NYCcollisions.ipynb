```python
import pandas as pd
import matplotlib  
import matplotlib.pyplot as plt  
from matplotlib import rcParams 
pd.options.display.mpl_style = 'default' #Better Styling  
new_style = {'grid': False} #Remove grid  
matplotlib.rc('axes', **new_style)  
rcParams['figure.figsize'] = (17.5, 17) #Size of figure  
rcParams['figure.dpi'] = 250

P=df_longlat.plot(kind='scatter', x='LONGITUDE', y='LATITUDE',color='white',xlim=(-74.06,-73.77),ylim=(40.61, 40.91),s=3.5,alpha=.4)
P.set_axis_bgcolor('black')
P.text(-74.050,40.885, 'All Collisions', color='white', fontsize=30)
#P.annotate('All Collisions', xytext=(-74.05, 40.89))
#fig = P.get_figure()
#fig.savefig('all_collisions.png')
```

```python
df_longlat['TIME'] = pd.to_datetime(df_longlat['TIME'])
acc_time = pd.DatetimeIndex(df_longlat['TIME'])
interv_mins = 60
df_longlat['TIME_CAT'] = (acc_time.hour * 60 + acc_time.minute) // interv_mins
df_longlat.ix[0:5,[0,1,2,3,4,df_longlat.shape[1]-1]]
```

```python
for iHour in range(24):
    P=df_longlat[df_longlat['TIME_CAT']==iHour].plot(kind='scatter', x='LONGITUDE', y='LATITUDE',color='white',xlim=(-74.06,-73.77),ylim=(40.61, 40.91),s=3.5,alpha=.4)
    P.set_axis_bgcolor('black')
    P.text(x=-74.050, y=40.885, color='white', fontsize=30,
           s='24H TIME: '+("%02d:%02d" % (iHour*interv_mins//60,iHour*interv_mins%60)))
    fig = P.get_figure()
    fig.savefig('collisions_hour'+("%02d" % iHour)+'.png')
    plt.close(fig)
```
